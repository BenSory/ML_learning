# ML : Regression Simple

```python
#Importation des packages de l'environnement de travail
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#Importation du jeu de données
df = pd.read_csv(r"C:/Users/maserati/Downloads/homeprices.csv")
df.head()
```

```python
#Visualisation du prix des appartements en fonction de leurs surfaces
%matplotlib inline
plt.xlabel("area(sqr.ft)")
plt.ylabel("price(US$)")
plt.scatter(df.area,df.price, color ="red", marker="+")
```

```python
#Spécification du modèle 
reg = linear_model.LinearRegression()
reg.fit(df[["area"]], df.price)
```

```python
#Prédiction du prix d'un appartement de 3300 pieds
reg.predict([[3300]])
```

```python
#coefficient
reg.coef_
```

```python
#Intercept
reg.intercept_
```

```python
# Importation des données pour futures prédiction
area_df = pd.read_csv(r"C:/Users/maserati/Downloads/areas.csv")
area_df.head(3)
```

```python
#Prédiction des prix
p=reg.predict(area_df)
p
```

```python
#Visualisation du prix des appartements en fonction de leurs surfaces avec l'intercept
%matplotlib inline
plt.xlabel("area(sqr.ft)")
plt.ylabel("price(US$)")
plt.scatter(df.area,df.price, color ="red", marker="+")
plt.plot(df.area ,reg.predict(df[["area"]]), color="blue")
```

# Exercice PIB Canada

```python
# Importation du jeu de données
canada = pd.read_csv(r"C:/Users/maserati/Downloads/canada_per_capita_income.csv")
canada.head(5)
```

```python
# Visualisation du PIB par tete canadien de 1970 à 2010
%matplotlib inline
plt.xlabel("year")
plt.ylabel("per capita income (US$)")

plt.plot(canada["year"], canada["per capita income (US$)"])
```

```python
#Modele de regression
reg1 = linear_model.LinearRegression()
reg1.fit(canada[["year"]],canada["per capita income (US$)"])
```

```python
#coef
reg1.coef_
```

```python
#Intercept
reg1.intercept_
```

```python
# Prédiction du PIB par tete en 2020
reg1.predict([[2020]])
```
